```python
import pandas as pd
import numpy as np
```

```python
CE = pd.read_csv("genage_models_export_CE_13062014.tsv",index_col=False,delimiter="\t")
SC = pd.read_csv("genage_models_export_SC_13062014.tsv",index_col=False,delimiter="\t")
DM = pd.read_csv("genage_models_export_DM_13062014.tsv",index_col=False,delimiter="\t")
MM = pd.read_csv("genage_models_export_MM_13062014.tsv",index_col=False,delimiter="\t")
gene2go = pd.read_csv("gene2go_13062014",index_col=False,delimiter="\t")
```

```python
MM["Longevity Influence"].value_counts()
```

```python
df = MM[["Entrez Gene ID","Longevity Influence"]].copy()
#在当前模型有机体中出现的GO_ID,将它转化为特征列
featureName = gene2go[gene2go["GeneID"].isin(df["Entrez Gene ID"].tolist())]["GO_ID"].drop_duplicates()
for name in featureName:
    df[name] = 0
#删除列，标签替换为01
df["Longevity Influence"].replace("anti",0,inplace=True)
df["Longevity Influence"].replace("pro",1,inplace=True)
df.drop(index=df[df["Longevity Influence"] == "fitness"].index.tolist(),inplace=True)
df.dropna(axis=0,subset=["Longevity Influence"],inplace=True)
#对每条数据对应的GO项赋值1
for index in df.index: 
    GeneID = df.loc[index,"Entrez Gene ID"]
    GOs = gene2go[gene2go["GeneID"] == GeneID]["GO_ID"]
    for GO in GOs:
        df.loc[index,GO] = 1
#对被注解少于3次的GO特征值过滤
# for c in df.columns[3:]:
#     if(df[c].sum() < 3):
#         df.drop(c,axis=1,inplace=True)
```

```python
df.to_csv("SC.csv")
```
